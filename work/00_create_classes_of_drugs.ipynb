```python
from tqdm import tqdm
import pandas as pd
import numpy as np
```

```python
ROOT_FOLDER = "data/"
```

```python
XLSX_INPUT_FILE = ROOT_FOLDER+"data_in.xlsx"
XLSX_OUTPUT_FILE = ROOT_FOLDER+"output_data_c.xlsx"
XLSX_OUTPUT_FILE_OTHERS = ROOT_FOLDER+"output_data_c_others.xlsx"
MAIN_EXCEL_SHEET = "Sheet_1"
```

- toate C-urile intre ele (C1 cu C2-C8, C2 cu C3-C8, etc)
- toate C-urile cu fiecare B
- toate C-urile cu fiecare M1, and so on.

```python
df = pd.read_excel(XLSX_INPUT_FILE, sheet_name = None)
```

```python
excel_sheets = list(df.keys())
print(excel_sheets)
```

```python
drug_pairs_c = []
drug_pairs_c_others = []
```

```python
#Get C pairs
for i in range(1,len(excel_sheets)):
    for j in range(i+1, len(excel_sheets)):
        sheet_name_a = excel_sheets[i]
        sheet_name_b = excel_sheets[j]
        if "C" in sheet_name_a and "C" in sheet_name_b:
            print(f"Checking {sheet_name_a} and {sheet_name_b}")
            df_a = df[sheet_name_a].replace(u'\xa0', u'', regex=True)
            df_b = df[sheet_name_b].replace(u'\xa0', u'', regex=True)
            for index1, row1 in df_a.iterrows():
                drug_df_a = row1['DrugName']
                for index2, row2 in df_b.iterrows():
                    drug_df_b = row2['DrugName']
                    drug_pairs_c.append((sheet_name_a, drug_df_a, sheet_name_b, drug_df_b,0,0,0,0))            
        
```

```python
output_df = pd.DataFrame(drug_pairs_c, columns =['ClassA', 'DrugA', 'ClassB', 'DrugB', 'DrugBankScore', 'DrugsScore', 'MedscapeScore','WebmdScore'])
```

```python
output_df
```

```python
output_df.to_excel(XLSX_OUTPUT_FILE,
             sheet_name='Sheet_name_1')  
```

```python
#Get Cs and others
for i in range(1,len(excel_sheets)):
    for j in range(i+1, len(excel_sheets)):
        sheet_name_a = excel_sheets[i]
        sheet_name_b = excel_sheets[j]
        if "C" in sheet_name_a and not "C" in sheet_name_b:
            print(f"Checking {sheet_name_a} and {sheet_name_b}")
            df_a = df[sheet_name_a].replace(u'\xa0', u'', regex=True)
            df_b = df[sheet_name_b].replace(u'\xa0', u'', regex=True)
            for index1, row1 in df_a.iterrows():
                drug_df_a = row1['DrugName']
                for index2, row2 in df_b.iterrows():
                    drug_df_b = row2['DrugName']
                    drug_pairs_c_others.append((sheet_name_a, drug_df_a, sheet_name_b, drug_df_b,0,0,0,0))            
        
```

```python
output_df_c_others = pd.DataFrame(drug_pairs_c_others, columns =['ClassA', 'DrugA', 'ClassB', 'DrugB', 'DrugBankScore', 'DrugsScore', 'MedscapeScore','WebmdScore'])
```

```python
output_df_c_others
```

```python
output_df_c_others.to_excel(XLSX_OUTPUT_FILE_OTHERS,
             sheet_name='Sheet_name_1')  
```
